```python
import tkinter as tk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg as fcta
from matplotlib.figure import Figure as figure

ppm_array = np.array([])
cocn_array = np.array([])
canvas = None
res_l=None

df_carbon = pd.DataFrame(columns=['Blank', 'Titrated Value', 'Weight', 'Percentage OC'])
df_nitrogen = pd.DataFrame(columns=['Volume of Extract', 'Titrated Value', 'Blank', 'Weight', 'Volume of Extract Taken', 'Percentage N'])
df_phosphorus = pd.DataFrame(columns=['PPM', 'Concentration'])

# initiating main window
root = tk.Tk(className='Soil Calculator')
label = tk.Label(root, text='Integrated Interface To Collect Data For Soil Calculator', font=("Helvetica", 16)).grid(row=1, column=0, sticky='s')
root.geometry('800x600')


# Define button functions
def submit():
    volume_K2Cr2O7.grid(row=9, column=0)
    K2Cr2O7_entry.grid(row=9, column=1)
    volume_FeSO4.grid(row=10, column=0)
    FeSO4_entry.grid(row=10, column=1)
    carbon_f_sub_but.grid(row=11, column=1)

def submiton():
    def f_function():
        S1 = 1
        V1 = volume_of_K2Cr2O7_EntryVar.get()
        V2 = volume_of_FeSO4_entryVar.get()
        if V2 == 0:
            tk.messagebox.showerror("Input Error", "Volume of FeSO4 cannot be zero.")
            return None
        S2 = (S1 * V1) / V2
        return S2
    
    f = f_function()
    if f is None:
        return
    
    b = carbon_blank_EntryVar.get()
    t = carbon_T_EntryVar.get()
    w = carbon_W_EntryVar.get()
    B = float(b)
    T = float(t)
    W = float(w)
    if W == 0:
        tk.messagebox.showerror("Input Error", "Weight cannot be zero.")
        return

    C = 0.39
    percentage_OC = ((B - T) * f * C) / W
    print('percentage_OC is ', percentage_OC)
    res_l = tk.Label(root, text=f"% of carbon is {percentage_OC}", font=('Times New Roman', 12, 'bold')).grid(row=14, column=0)

    volume_of_K2Cr2O7_EntryVar.set("")
    volume_of_FeSO4_entryVar.set("")
    carbon_blank_EntryVar.set("")
    carbon_T_EntryVar.set("")
    carbon_W_EntryVar.set("")

def submit_Ndata():
    choice_f_known.grid(row=9, column=0)
    choice_f_unknown.grid(row=9, column=1)

def toggel():
    known_f_label.grid(row=10, column=0)
    known_f_entry.grid(row=10, column=1)
    sub_but_known_f.grid(row=11, column=0)

def calculate_with_known_f():
    try:
        v1_ = float(v1_EntryVar.get())
        tn_ = float(Tn_EntryVar.get())
        bn_ = float(Bn_EntryVar.get())
        wn_ = float(Wn_EntryVar.get())
        v2_ = float(v_et_EntryVar.get())

        if wn_ == 0 or v2_ == 0:
            tk.messagebox.showerror("Input Error", "Weight (Wn) and Volume of Extract Taken (v2) cannot be zero.")
            return

        S2 = float(known_f.get())
        per_N = ((0.014 * S2 * v1_ * (tn_ - bn_)) / (wn_ * v2_))
        tk.Label(root, text=f"% of nitrogen is {per_N}", font=('Times New Roman', 12, 'bold')).grid(row=14, column=0)

        v1_EntryVar.set('')
        Tn_EntryVar.set('')
        Bn_EntryVar.set('')
        Wn_EntryVar.set('')
        v_et_EntryVar.set('')
        known_f.set('')
    except ValueError:
        tk.messagebox.showerror("Input Error", "Please enter valid numeric values.")

def toggel_2():
    choice_f_known.grid_remove()
    choice_f_unknown.grid_remove()
    k_l.grid(row=10, column=0)
    k_entry.grid(row=10, column=1)
    f_l.grid(row=11, column=0)
    f_entry.grid(row=11, column=1)
    sub_but.grid(row=13, column=1)

def inside_toggel():
    try:
        v1_ = float(v1_EntryVar.get())
        tn_ = float(Tn_EntryVar.get())
        bn_ = float(Bn_EntryVar.get())
        wn_ = float(Wn_EntryVar.get())
        v2_ = float(v_et_EntryVar.get())
        S1 = 1
        V1 = float(v_K2Cr2O7.get())
        V2 = float(v_FeSO4.get())

        if V2 == 0:
            tk.messagebox.showerror("Input Error", "Volume of FeSO4 cannot be zero.")
            return

        S2 = (S1 * V1) / V2
        if wn_ == 0 or v2_ == 0:
            tk.messagebox.showerror("Input Error", "Weight (Wn) and Volume of Extract Taken (v2) cannot be zero.")
            return

        per_N = ((0.014 * S2 * v1_ * (tn_ - bn_)) / (wn_ * v2_))
        tk.Label(root, text=f"% of nitrogen is {per_N}", font=('Times New Roman', 12, 'bold')).grid(row=16, column=0)

        v1_EntryVar.set('')
        Tn_EntryVar.set('')
        Bn_EntryVar.set('')
        Wn_EntryVar.set('')
        v_et_EntryVar.set('')
        k_entry.set("")
        f_entry.set("")
    except ValueError:
        tk.messagebox.showerror("Input Error", "Please enter valid numeric values.")

def append_entry():
    global ppm_array, cocn_array
    try:
        ppm_value = float(ppm_var.get())
        cocn_value = float(cocn_var.get())

        ppm_array = np.append(ppm_array, ppm_value)
        cocn_array = np.append(cocn_array, cocn_value)

        ppm_var.set('')
        cocn_var.set('')
    except ValueError:
        tk.messagebox.showerror("Input Error", "Please enter valid numeric values for PPM and Concentration.")

def gen_curve():
    global ppm_array, cocn_array, canvas
    if canvas:
        canvas.get_tk_widget().destroy()

    fig, ax = plt.subplots()
    ax.plot(ppm_array, cocn_array, marker='o', linestyle='-', color='b')
    ax.set_xlabel('PPM')
    ax.set_ylabel('Concentration')
    ax.set_title('PPM vs Concentration')
    ax.grid(True)

    canvas = fcta(fig, master=root)
    canvas.draw()
    canvas.get_tk_widget().grid(row=18, column=0, columnspan=2)

def clear_curve():
    global canvas
    if canvas:
        canvas.get_tk_widget().destroy()
        canvas = None


def cal_CEC():
    pass

def cal_PSA():
    pass

def cal_Enzymes():
    pass
        
        
        

def hide_new_widgets():
    widgets = [choice_f_known, choice_f_unknown, k_entry, k_l, f_entry, f_l, sub_but, known_f_label, known_f_entry, sub_but_known_f]
    for widget in widgets:
        widget.grid_remove()

def hide_all_widgets():
    widgets = [carbon_Blank_label, entry1, carbon_T_label, entry2, carbon_W_label, entry3, carbon_sub_but,
               volume_K2Cr2O7, K2Cr2O7_entry, volume_FeSO4, FeSO4_entry, carbon_f_sub_but,
               v_e_label, entry4, Bn_label, entry5, Tn_label, entry6, Wn_label, entry7, v_et_label, entry8, sub_but2,
               cal_cur_label, cal_cur_label2, ppm_entry, cocn_entry, append_but, submit_p_butt, clear_curve_but]
    for widget in widgets:
        widget.grid_remove()
    hide_new_widgets()
def hide_result_label():
    if res_l in globals() != None:
        res_l.grid_remove()
        
def cal_carbon():
    hide_all_widgets()
    hide_result_label()
    carbon_Blank_label.grid(row=2, column=0)
    entry1.grid(row=2, column=1)
    carbon_T_label.grid(row=3, column=0)
    entry2.grid(row=3, column=1)
    carbon_W_label.grid(row=4, column=0)
    entry3.grid(row=4, column=1)
    carbon_sub_but.grid(row=6, column=1)

def cal_nitrogen():
    hide_all_widgets()
    hide_result_label()
    v_e_label.grid(row=2, column=0)
    entry4.grid(row=2, column=1)
    Bn_label.grid(row=3, column=0)
    entry5.grid(row=3, column=1)
    Tn_label.grid(row=4, column=0)
    entry6.grid(row=4, column=1)
    Wn_label.grid(row=5, column=0)
    entry7.grid(row=5, column=1)
    v_et_label.grid(row=6, column=0)
    entry8.grid(row=6, column=1)
    sub_but2.grid(row=7, column=1)

def cal_phosphorus():
    hide_all_widgets()
    hide_result_label()
    cal_cur_label.grid(row=2, column=0)
    ppm_entry.grid(row=2, column=1)
    cal_cur_label2.grid(row=3, column=0)
    cocn_entry.grid(row=3, column=1)
    append_but.grid(row=5, column=0)
    submit_p_butt.grid(row=5, column=1)
    clear_curve_but.grid(row=5, column=2)

def data_frame():
    global df
    df = pd.DataFrame({
        'PPM': ppm_array,
        'Concentration': cocn_array
    })
    text_widget = tk.Text(root, height=20, width=100)
    text_widget.grid(row=19, column=0, columnspan=2)
    text_widget.insert(tk.END, df.to_string())
    text_widget.insert(tk.END, df_nitrogen.to_string())

def save_df():
    global df
    df.to_excel('C:\\Users\\CP\\Desktop\\output.xlsx')

# Creating menu bar
menubar = tk.Menu(root)

# Adding calculation menu
calculation = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='Calculation', menu=calculation)
calculation.add_command(label='Carbon', command=cal_carbon)
calculation.add_command(label='Nitrogen', command=cal_nitrogen)
calculation.add_command(label='Phosphorus', command=cal_phosphorus)
calculation.add_command(label = 'CEC',command =cal_CEC)
calculation.add_command(label = 'Particle size analysis',command = cal_PSA)
calculation.add_command(label = 'Enzyme',command = cal_Enzymes)
# Adding view option in menubar
view = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='View', menu=view)
view.add_command(label='Data', command=data_frame)
view.add_command(label='Save', command=save_df)

# Carbon calculation entry widgets
carbon_blank_EntryVar = tk.DoubleVar()
carbon_T_EntryVar = tk.DoubleVar()
carbon_W_EntryVar = tk.DoubleVar()
volume_of_K2Cr2O7_EntryVar = tk.DoubleVar()
volume_of_FeSO4_entryVar = tk.DoubleVar()

carbon_Blank_label = tk.Label(root, text='Enter Blank', font=("Times New Roman", 14))
carbon_T_label = tk.Label(root, text='Enter Titrated Value', font=("Times New Roman", 14))
carbon_W_label = tk.Label(root, text='Enter Weight', font=("Times New Roman", 14))
volume_K2Cr2O7 = tk.Label(root, text='Volume of K2Cr2O7', font=("Times New Roman", 14))
volume_FeSO4 = tk.Label(root, text='Volume of FeSO4', font=("Times New Roman", 14))

entry1 = tk.Entry(root, textvariable=carbon_blank_EntryVar)
entry2 = tk.Entry(root, textvariable=carbon_T_EntryVar)
entry3 = tk.Entry(root, textvariable=carbon_W_EntryVar)
K2Cr2O7_entry = tk.Entry(root, textvariable=volume_of_K2Cr2O7_EntryVar)
FeSO4_entry = tk.Entry(root, textvariable=volume_of_FeSO4_entryVar)

carbon_sub_but = tk.Button(root, text='submit', command=submit)
carbon_f_sub_but = tk.Button(root, text='submit', command=submiton)

# Nitrogen calculation entry widgets
v1_EntryVar = tk.DoubleVar()
Bn_EntryVar = tk.DoubleVar()
Tn_EntryVar = tk.DoubleVar()
Wn_EntryVar = tk.DoubleVar()
v_et_EntryVar = tk.DoubleVar()

v_e_label = tk.Label(root, text='Enter Volume of Extract', font=("Times New Roman", 14))
Bn_label = tk.Label(root, text='Enter Blank', font=("Times New Roman", 14))
Tn_label = tk.Label(root, text='Enter Titrated Value', font=("Times New Roman", 14))
Wn_label = tk.Label(root, text='Enter Weight', font=("Times New Roman", 14))
v_et_label = tk.Label(root, text='Enter Volume of Extract Taken', font=("Times New Roman", 14))

entry4 = tk.Entry(root, textvariable=v1_EntryVar)
entry5 = tk.Entry(root, textvariable=Bn_EntryVar)
entry6 = tk.Entry(root, textvariable=Tn_EntryVar)
entry7 = tk.Entry(root, textvariable=Wn_EntryVar)
entry8 = tk.Entry(root, textvariable=v_et_EntryVar)

sub_but2 = tk.Button(root, text='Submit', command=submit_Ndata)
choice_f_known = tk.Button(root, text='I know f value', command=toggel)
choice_f_unknown = tk.Button(root, text='I don\'t know f value', command=toggel_2)

known_f = tk.DoubleVar()
known_f_label = tk.Label(root, text='Enter f value')
known_f_entry = tk.Entry(root, textvariable=known_f)
sub_but_known_f = tk.Button(root, text='submit', command=calculate_with_known_f)

v_K2Cr2O7 = tk.DoubleVar()
k_l = tk.Label(root, text='Volume of K2Cr2O7')
k_entry = tk.Entry(root, textvariable=v_K2Cr2O7)

v_FeSO4 = tk.DoubleVar()
f_l = tk.Label(root, text='Volume of FeSO4')
f_entry = tk.Entry(root, textvariable=v_FeSO4)

sub_but = tk.Button(root, text='submit', command=inside_toggel)

# Phosphorus calculation entry widgets
ppm_var = tk.DoubleVar()
cocn_var = tk.DoubleVar()

cal_cur_label = tk.Label(root, text='Enter PPM', font=("Times New Roman", 14))
cal_cur_label2 = tk.Label(root, text='Enter Concentration', font=("Times New Roman", 14))
ppm_entry = tk.Entry(root, textvariable=ppm_var)
cocn_entry = tk.Entry(root, textvariable=cocn_var)

append_but = tk.Button(root, text='Append', command=append_entry)
submit_p_butt = tk.Button(root, text='Generate Curve', command=gen_curve)
clear_curve_but = tk.Button(root, text='Clear Curve', command=clear_curve)

# Running the Tkinter event loop
root.config(menu=menubar)
root.mainloop()
```
